#Task 5 - Build ML model with Linear Regression (Target column is price)

```python
df=df[['price','sqft_living']]
```

```python
df
```

```python
# Independent Variable
```

```python
X=df['sqft_living']
```

```python
# Dependent variable
```

```python
Y=df['price']
```

```python
# Main process
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X,Y)
plt.xlabel('Sqft-living')
plt.ylabel('Price')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.4,random_state=23)
```

```python
X_train=np.array(X_train).reshape(-1,1)
```

```python
X_train
```

```python
X_test=np.array(X_test).reshape(-1,1)
```

```python
X_test
```

```python
# Fitting of Linear Regression
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(X_train,Y_train)
```

```python
C=lr.intercept_
C
```

```python
M=lr.coef_
M
```

Since we have M,C values lets try predicting in Y=MX+C

```python
Y_pred_train=M*X_train+C
Y_pred_train.flatten()
```

```python
Y_pred_train1=lr.predict(X_train)
Y_pred_train1
```

```python
plt.scatter(X_train,Y_train)
plt.scatter(X_train,Y_pred_train1,color='red')
plt.xlabel('Sqft-living')
plt.ylabel('Price')
```

```python
plt.scatter(X_train,Y_train)
plt.plot(X_train,Y_pred_train1,color='red')
plt.xlabel('Sqft-living')

plt.ylabel('Price')
```

Now let's try fitting the model for X_test, Y_test

```python
lr.fit(X_test,Y_test)
```

```python
Y_pred_test=M*X_test+C
Y_pred_test.flatten()
```

```python
Y_pred_test1=lr.predict(X_test)
Y_pred_test1
```

```python
plt.scatter(X_test,Y_test)
plt.scatter(X_test,Y_pred_test1,color='red')
plt.xlabel('Sqft-living')
plt.ylabel('Price')
```

```python
plt.scatter(X_test,Y_test)
plt.plot(X_test,Y_pred_test1,color='red')
plt.xlabel('Sqft-living')
plt.ylabel('Price')
```

Therefore the linear model is satisfied for both train and test data.
